```python
# Convert to pdbs
import os

import incito_pipeline.util.parse.yaml_parse as yaml_p
import incito_pipeline.util.parse.pdb as pdb_p
import incito_pipeline.util.data_utility as du

path_to_boltz_output = "/spinning1/home/cp864/boltz_2_new/19_08_2025/output/seperated/fv"
output_dir = "/spinning1/home/cp864/boltz_2_new/19_08_2025/output/seperated/fv_pdbs"

dirs = du.get_dirs(path_to_boltz_output)

for dir in dirs:
    dir_name = os.path.basename(dir).split('boltz_results_')[-1]

    path_to_predictions = os.path.join(path_to_boltz_output, dir, "predictions")
    save_dir = os.path.join(output_dir, dir_name)

    ab_path = os.path.join(path_to_predictions, "ab/ab_model_0.cif")
    ag_path = os.path.join(path_to_predictions, "ag/ag_model_0.cif")

    os.makedirs(save_dir, exist_ok=True)

    pdb_p.pdb_parse(ab_path, os.path.join(save_dir, "ab.pdb"))
    pdb_p.pdb_parse(ag_path, os.path.join(save_dir, "ag.pdb"))


```
