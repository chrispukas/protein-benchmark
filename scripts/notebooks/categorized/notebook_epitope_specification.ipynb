```python
# Converting .cif files to .pdb files:
import os
import incito_pipeline.util.parse.pdb as pdb_p

generic_src_dir = "/spinning1/home/cp864/boltz_2_new/19_08_2025/output"

dir_list = [("specification_fixed/fv/boltz_results_fv/predictions", "specification_fixed/fv/pdb"),
            ("specification_fixed/vhh/boltz_results_vhh/predictions", "specification_fixed/vhh/pdb"),]

for src, dest in dir_list:
    src = os.path.join(generic_src_dir, src)
    dest = os.path.join(generic_src_dir, dest)

    os.makedirs(dest, exist_ok=True)
    print(f"Converting {src} -> {dest}")
    
    pdb_p.convert_pdb_from_cif_dataset(src, dest, override=False, simplify_name=True)
```

# Benchmark for epitope specification

```python
import incito_pipeline.util.dockq_utility as dqu
import incito_pipeline.util.data_utility as du
import incito_pipeline.util.plotting as plt

import incito_pipeline.util.cache as c

stratized = {}

def dq_fn(path_1: str, 
          path_2: str):
    pairs = du.pairs_pdbs(path_1, path_2)
    full_results, dockq_scores, failed_count, names = dqu.get_dockq_score_from_pairs(pairs)
    return full_results, dockq_scores, failed_count, names

fv_generated_pdbs_dir = "/spinning1/home/cp864/boltz_2_new/19_08_2025/output/specification_fixed/fv/pdb"
fv_original_pdbs_dir = "/home/cp864/repos/incito-pipeline/incito_pipeline/datasets/AF3_independent_test_Fv_GroundTruth"

full_results, dockq_scores, failed_count, names = c.c_template("fv_dockq_scores_epitope_specification_fixed", dq_fn, fv_original_pdbs_dir, fv_generated_pdbs_dir)
print(f"{len(dockq_scores)} dockq scores calculated, {failed_count} failed")
stratized[f"fv dockq scores #{len(dockq_scores)}"] = du.stratize_data(dockq_scores)

vhh_generated_pdbs_dir = "/spinning1/home/cp864/boltz_2_new/19_08_2025/output/specification_fixed/vhh/pdb"
vhh_original_pdbs_dir = "/home/cp864/repos/incito-pipeline/incito_pipeline/datasets/AF3_independent_test_VHHs_GroundTruth"

full_results, dockq_scores, failed_count, names = c.c_template("vhh_dockq_scores_epitope_specification_fixed", dq_fn, vhh_original_pdbs_dir, vhh_generated_pdbs_dir)
print(f"{len(dockq_scores)} dockq scores calculated, {failed_count} failed")
stratized[f"vhh dockq scores #{len(dockq_scores)}"] = du.stratize_data(dockq_scores)

plt.plot_stacked_bar_chart(stratized, title="DockQ scores for epitope specification")

zipped = zip(dockq_scores, names)
print(sorted(zipped))
```
